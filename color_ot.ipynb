# Color transfer by optimal transport

```python
import os

import numpy as np
import matplotlib.pyplot as plt

from PIL import Image

# super pixels
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
```

```python
# image loading
flower1 = Image.open('flower_1.jpg')
flower2 = Image.open('flower_2.jpg')

# conversion to [0,1]-valued arrays
u = np.array(flower1, dtype='float') / 255
v = np.array(flower2, dtype='float') / 255

# image display
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.imshow(flower1)
ax1.axis('off')

ax2.imshow(flower2)
ax2.axis('off');
```

```python
# super pixels computation
num_segments = 200
u_segments = slic(u, n_segments=num_segments, sigma=5)
v_segments = slic(v, n_segments=num_segments, sigma=5)

# super pixels display
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.imshow(mark_boundaries(u, u_segments))
ax1.axis("off")

ax2.imshow(mark_boundaries(v, v_segments))
ax2.axis("off");
```
